# **Tarea Modelo no Supervisado**

## **Clustering**



```
import numpy as np
import pandas as pd 
from sklearn.cluster import MiniBatchKMeans
import seaborn as sns

candies = pd.read_csv("https://raw.githubusercontent.com/hanGtz92/candy/main/candy.csv")
x = candies.drop("nombre", axis = 1)
kmeans = MiniBatchKMeans(n_clusters = 4, batch_size = 8).fit(x) 

candies['cluster'] = kmeans.predict(x)
candies
```

```
sns.pairplot(candies[['p_azucar','p_precio','p_ganador','cluster']], hue = 'cluster')
```

```
sns.scatterplot(data = candies, x = "p_azucar", y = "p_ganador", hue = "cluster", palette = "deep")
```
